```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Problem 2

```python
t_s = np.linspace(0,5, num=100)
X = np.array([0, 1.3, 2.1, 2.8, 4.2, 5.7])
y = np.array([-1, -1, -1, 1, -1, 1], dtype=np.float32)
z_s = X[:,None] - t_s[None,:]
J_s = np.sum(np.maximum(0, 1 - z_s*y[:,None]), axis=0)

plt.figure(figsize=(10,5))
plt.plot(t_s, J_s)
plt.xlabel("t value")
plt.ylabel("Hinge Loss")
plt.grid()

arg_min = np.argmin(J_s)
print(arg_min)
print(f"One value of t that minimizes the Hinge Loss is: {t_s[arg_min]}")
print(np.min(J_s))
print(J_s[70]) # the are several values that minimize the Hinge loss
```

```python
t_min = t_s[arg_min]
z = X - t_min
e_i = np.maximum(0, 1 - z*y)

print(e_i.shape)
print(e_i)
```

```python
t_min = 4.4
z = X - t_min
e_i = np.maximum(0, 1 - z*y)

print(e_i.shape)
print(e_i)
print(f"Loss: {np.sum(e_i)}")
```

Problem 4

```python
X = np.array([0,1,2,3])
y = np.array([1,-1,1,-1])
x = np.linspace(-2,5,100)
gam1 = 3.0
alpha1 = np.array([0,0,1,1])
gam2 = 0.3
alpha2 = np.array([1,1,1,1])
```

```python
x_x = (x[:,None] - X[None,:])**2
K1 = np.exp(-gam1*x_x)
z1 = np.sum(alpha1*y*K1, axis=1)
yhat1 = 1*(z1>0) - 1*(z1<=0)

print(f"The number of missclassified samples in the first case is {1}")

plt.figure()
plt.plot(x,z1, label='z')
plt.plot(x,yhat1, label='yhat')
plt.legend()
plt.grid()
```

```python
K2 = np.exp(-gam2*x_x)
z2 = np.sum(alpha2*y*K2, axis=1)
yhat2 = 1*(z2>0) - 1*(z2<=0)
print(f"The number of missclassified samples in the second case is {2}")

plt.figure()
plt.plot(x,z2, label='z')
plt.plot(x,yhat2, label='yhat')
plt.legend()
plt.grid()
```

```python
# gamma = 3; alpha = [1 1 1 1]
K = np.exp(-gam1*x_x)
z = (K*y[None,:]).dot(alpha2)
yhat = 1*(z>0) - 1*(z<=0)

print(f"The number of missclassified samples in the second case is {0}")
plt.figure()
plt.plot(x,z)
plt.plot(x,yhat)
plt.grid()
```
