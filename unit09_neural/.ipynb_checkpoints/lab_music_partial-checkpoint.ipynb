Extracting features from audio files is an entire subject on its own right.  A commonly used set of features are called the Mel Frequency Cepstral Coefficients (MFCCs).  These are derived from the so-called mel spectrogram which is something like a regular spectrogram, but the power and frequency are represented in log scale, which more naturally aligns with human perceptual processing.  You can run the code below to display the mel spectrogram from the audio sample.

You can easily see the four notes played in the audio track.  You also see the 'harmonics' of each notes, which are other tones at integer multiples of the fundamental frequency of each note.

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
librosa.display.specshow(librosa.amplitude_to_db(S),
                         y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```

Using the MFCC features described above, Eric Humphrey and Juan Bellow have created a complete data set that can used for instrument classification.  Essentially, they collected a number of data files from the website above.  For each audio file, the segmented the track into notes and then extracted 120 MFCCs for each note. Note the password.  Load the four files into some directory, say  `instrument_dataset`.  Then, load them with the commands.

```python
data_dir = './instrument_dataset/'
Xtr = np.load(data_dir+'uiowa_train_data.npy')
ytr = np.load(data_dir+'uiowa_train_labels.npy')
Xts = np.load(data_dir+'uiowa_test_data.npy')
yts = np.load(data_dir+'uiowa_test_labels.npy')
```

Looking at the data files:
* What are the number of training and test samples?
* What is the number of features for each sample?
* How many classes (i.e. instruments) are there per class?


```python
# TODO
print(f"The number of training samples is: {Xtr.shape[0]}")
print(f"The number of training samples is: {Xts.shape[0]}")
print(f"The number of features for each sample is: {Xtr.shape[1]}")
n_classes = len(np.unique(ytr))
print(f"The number of classes si: {n_classes}")
```

Before continuing, you must scale the training and test data, `Xtr` and `Xts`.  Compute the mean and std deviation of each feature in `Xtr` and create a new training data set, `Xtr_scale`, by subtracting the mean and dividing by the std deviation.  Also compute a scaled test data set, `Xts_scale` using the mean and std deviation learned from the training data set.

```python
# TODO Scale the training and test matrices
Xtr_mean = np.mean(Xtr, axis=0)
Xtr_std = np.std(Xtr, axis=0)
Xtr_scale = (Xtr - Xtr_mean[None,:])/Xtr_std[None,:]
Xts_scale = (Xts - Xtr_mean[None,:])/Xtr_std[None,:]
```

## Building a Neural Network Classifier

Following the example in [MNIST neural network demo](./mnist_neural.ipynb), clear the keras session.  Then, create a neural network `model` with:
* `nh=256` hidden units
* `sigmoid` activation
* select the input and output shapes correctly
* print the model summary

```python
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import optimizers
import tensorflow.keras.backend as K
```

```python
# TODO clear session
K.clear_session()
```

```python
import tensorflow.keras.callbacks

class PeriodicPrintStats(tensorflow.keras.callbacks.Callback):
    def __init__(self, period=100):
        self.period = period
        
    def on_train_begin(self, logs={}):
        self.losses = []
        self.test_acc = []
        self.train_acc = []
        self.step = 0

    def on_batch_end(self, batch, logs={}):
        # Record Train loss and Train accuracy
        loss = logs.get('loss')
        acc = logs.get('accuracy')
        self.losses.append(loss)
        self.train_acc.append(acc)
        
        if (self.step % self.period == 0):
            print(f"Step: {self.step}; Loss: {loss}; Train acc: {acc}")
        self.step += 1
    
    def on_epoch_end(self, epoch, logs={}):
        val_acc = logs.get('val_accuracy')
        self.test_acc.append(val_acc)
        print(f"The validation accuracy on test data is: {val_acc}")
        
```

```python
# TODO: construct the model
in_dim = Xtr_scale.shape[1]
out_dim = n_classes
nh = 256
model = Sequential()
model.add(Dense(units=nh, input_shape=(in_dim,), activation='sigmoid', name='hidden'))
model.add(Dense(units=out_dim, activation='softmax', name='output'))
```

```python
# TODO:  Print the model summary
model.summary()
```

Create an optimizer and compile the model.  Select the appropriate loss function and metrics.  For the optimizer, use the Adam optimizer with a learning rate of 0.001

```python
# TODO
opt = optimizers.Adam(lr=0.001)
model.compile(optimizer=opt,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Fit the model for 10 epochs using the scaled data for both the training and validation.  Use the `validation_data` option to pass the test data.  Also, pass the callback class create above.  Use a batch size of 100.  Your final accuracy should be >99%.

```python
# TODO
n_epochs = 10
batch_size = 100
period = 200
print_callb = PeriodicPrintStats(period=period)
hist = model.fit(Xtr_scale, ytr, batch_size=batch_size, epochs=n_epochs, verbose=0,
                 callbacks=[print_callb], validation_data=(Xts_scale, yts))
```

Plot the validation accuracy saved in `hist.history` dictionary. This gives one accuracy value per epoch.  You should see that the validation accuracy saturates at a little higher than 99%.  After that it "bounces around" due to the noise in the stochastic gradient descent.

```python
# TODO
plt.figure(figsize=(10,5))
# using the data stored in the callback object which is the same as in hist.history['val_accuracy']
plt.plot(np.arange(n_epochs)+1, print_callb.test_acc, label='Validation Accuracy')
plt.grid()
plt.xlim(1,10)
plt.legend()
plt.title("Validation Accuracy vs Epoch", fontsize=12)
plt.ylabel("Validation Accuracy", fontsize=12)
plt.xlabel("Epoch", fontsize=12)

plt.figure(figsize=(10,5))
plt.plot(np.arange(len(print_callb.train_acc)), print_callb.train_acc, label='Training Accuracy')
plt.grid()
plt.legend()
plt.title("Training Accuracy vs Step", fontsize=12)
plt.ylabel("Training Accuracy", fontsize=12)
plt.xlabel("Step", fontsize=12)
```

Plot the loss values saved in the `hist.history` dictionary.  You should see that the loss is steadily decreasing.  Use the `semilogy` plot.

```python
# TODO
plt.figure(figsize=(10,5))
plt.semilogy(np.arange(n_epochs)+1, hist.history['loss'], label='Loss')
plt.grid()
plt.xlim(1,10)
plt.legend()
plt.title("Loss vs Epoch: log-scale", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.xlabel("Epoch", fontsize=12)
```

## Optimizing the Learning Rate

One challenge in training neural networks is the selection of the learning rate.  Rerun the above code, trying four learning rates as shown in the vector `rates`.  For each learning rate:
* clear the session
* construct the network
* select the optimizer.  Use the Adam optimizer with the appropriate learrning rate.
* train the model for 20 epochs
* save the accuracy and losses

```python
rates = [0.01,0.001,0.0001]
batch_size = 100
loss_hist = [None]*len(rates)
val_acc = [None]*len(rates)

# TODO
for i,lr in enumerate(rates):
    K.clear_session()
    model = Sequential()
    model.add(Dense(units=nh, input_shape=(in_dim,), activation='sigmoid', name='hidden'))
    model.add(Dense(units=out_dim, activation='softmax', name='output'))
    opt = optimizers.Adam(lr=lr)
    model.compile(optimizer=opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    hist = model.fit(Xtr_scale, ytr, batch_size=batch_size, epochs=20, verbose=0, validation_data=(Xts_scale, yts))
    loss_hist[i] = hist.history['loss']
    val_acc[i] = hist.history['val_accuracy']
```

Plot the loss funciton vs. the epoch number for all three learning rates on one graph.  You should see that the lower learning rates are more stable, but converge slower.

```python
# TODO
plt.figure(figsize=(10,5))
for i,lr in enumerate(rates):
    plt.plot(np.arange(20)+1, loss_hist[i], label='Lr='+str(lr))
plt.grid()
plt.xlim(1,20)
plt.legend()
plt.title("Loss vs Epoch: for three different Learning rates", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.xlabel("Epoch", fontsize=12)
plt.xticks(np.arange(1,21), np.arange(1,21))

plt.figure(figsize=(10,5))
for i,lr in enumerate(rates):
    plt.plot(np.arange(20)+1, val_acc[i], label='Lr='+str(lr))
plt.grid()
plt.xlim(1,20)
plt.legend()
plt.title("Validation Accuracy vs Epoch: for three different Learning rates", fontsize=12)
plt.ylabel("Validation Accuracy", fontsize=12)
plt.xlabel("Epoch", fontsize=12)
plt.xticks(np.arange(1,21), np.arange(1,21))
```

From the figures above, it's clear that a lower learning rate is more stable but the convergence is slower.
