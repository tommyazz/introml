```python
#exam
x = np.array([[1,1],[-1,-1]])
y = np.array([1,1])
Wh = np.array([[1,0],[0,1],[-1,1]])
print(Wh.shape)
bh = np.array([-0.5,-1,3])
```

```python
Z = Wh.dot(x.T) + bh[:,None]
print(Z)
```

```python
Z = np.sum(x[:,None,:]*Wh[None,:,:], axis=2) + bh[None,:]
print(Z)
U = np.maximum(0,Z)
#print(U)
Utr = np.hstack([U, np.ones((U.shape[0],))[:,None]])
print(Utr)
coeff = np.linalg.lstsq(Utr, y, rcond=None)[0]
print(coeff)
Wo = coeff[:-1]
bo = coeff[-1]
yhat = U.dot(Wo) + bo
print(yhat)
```

```python
xtr = np.array([-2,-1,0,3,3.5])
ytr = np.array([0,0,1,3,3])
u_tr = np.maximum(0, xtr[:,None]*Wh[None,:] + bh[None,:])
print(u_tr.shape)
U = np.hstack([u_tr, np.ones(u_tr.shape[0])[:,None]])
coeff = np.linalg.lstsq(U, ytr, rcond=None)[0]
```

```python
print(coeff)
```

```python
Wo = coeff[:-1]
bo = coeff[-1]
```

```python
print(Wo)
print(bo)
```

```python
def predict(x,Wh,bh,Wo,bo):
    z = x[:,None]*Wh[None,:] + bh[None,:]
    u = np.maximum(0, z)
    yhat = u.dot(Wo) + bo
    return yhat
```

```python
yhat_tr = predict(xtr,Wh,bh,Wo,bo)
plt.figure(figsize=(10,5))
plt.scatter(ytr, yhat_tr)
plt.grid()
print(f"MSE_tr={np.mean((ytr - yhat_tr)**2)}")
```

```python

xts = np.linspace(-3,3,100)
yhat = predict(xts,Wh,bh,Wo,bo)
print(yhat.shape)
plt.figure(figsize=(12,6))
plt.scatter(xts, yhat)
plt.grid()
plt.xlabel('x')
plt.ylabel('yhat')
plt.title('yhat vs x')
```

Problem 3

```python
nsamp = 400   # number of samples to generate
nin = 2        # number of dimensions of each sample
X = np.random.uniform(0,1,(nsamp,nin))
y = np.random.uniform(0,1,(nsamp,))
print(X.shape)
print(y.shape)
```

```python
def loss_eval(param, X, y):
    """
    Evaluates the loss function and gradients 
    for the neural network
    """
    
    # Unpack the parameters
    W, b, a = param
    
    # Hidden layer
    Zh = X.dot(W) + b[None,:]
    Uh = 1/(1+np.exp(-Zh))
    
    # Yhat
    sum_Uh = np.sum(Uh, axis=1)
    yhat = np.sum(Uh*a[None,:], axis=1)/sum_Uh
    
    # RSS
    loss = np.sum((y - yhat)**2)
    
    ######### This part implements dL_du as in question 3.h #########################
    yerr = yhat - y
    sum_Uh_2 = sum_Uh**2
    sum_a_Uh = Uh.dot(a)
    grad_yhat_du = (sum_Uh[:,None]*a[None,:] - sum_a_Uh[:,None])/(sum_Uh_2[:,None])
    grad_L_du = 2*yerr[:,None]*grad_yhat_du
    #################################################################################

    grad_u_z = Uh*(1-Uh)
    grad_L_dz = grad_L_du*grad_u_z 

    grad_Wh = np.sum(grad_L_dz[:,None,:]*X[:,:,None], axis=0)
    grad_bh = np.sum(grad_L_dz, axis=0)
    grad_a = np.sum((Uh/sum_Uh[:,None])*yerr[:,None], axis=0)*2
    
    # Pack the gradients
    grad = [grad_Wh, grad_bh, grad_a]
    
    return loss, grad
```

```python
# TODO
nh = 4 # number of hidden units
W = np.random.normal(0,1,(nin,nh))
b = np.random.normal(0,1,(nh))
a = np.random.normal(0,1,(nh))
param0 = [W,b,a]
param1 = []
step = 1e-6
for i, p0 in enumerate(param0):
    p1 = p0 + np.random.normal(0,step,p0.shape)
    param1.append(p1)
    
# Compute the loss and gradient at the parameters 
loss0, grad0 = loss_eval(param0, X, y)
loss1, grad1 = loss_eval(param1, X, y)

# Compute the expected change in loss from the gradient
dloss_est = 0
for i in range(len(grad0)):
    dloss_est += np.sum(grad0[i]*(param1[i]-param0[i]))
    
# Print the results    
print('Actual change in loss:    %12.4e' % (loss1-loss0))
print('Predicted change in loss: %12.4e' % dloss_est)
```

For testing the gradient computation, I assumed an input x with 2 features (Ni=2), and 4 hidden units (M=4) in the network
