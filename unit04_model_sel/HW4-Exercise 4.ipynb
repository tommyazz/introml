```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Xtr = np.linspace(0,1.0, num=10)
Xtr_new = np.linspace(1.0,2.0, num=10)
```

```python
beta = np.array([1,2,-1])
print(beta)
```

```python
Xtr_mean = np.mean(Xtr)
Xtr_new_mean = np.mean(Xtr_new)
def transform(X):
    n = len(X)
    c1 = np.ones(n,)
    c2 = X
    c3 = X**2
    return np.column_stack((c1,c2,c3))
```

```python
ytr = np.dot(transform(Xtr), beta)
ytr_mean = np.mean(ytr)
ytr_new = np.dot(transform(Xtr_new), beta)
ytr_new_mean = np.mean(ytr_new)
```

```python
beta_1 = np.sum((Xtr-Xtr_mean)*(ytr-ytr_mean)) / np.sum((Xtr-Xtr_mean)**2)
beta_0 = ytr_mean - beta_1*Xtr_mean
print(f"The estimated beta_0 is: {beta_0}; The estimated beta_1 is: {beta_1}")
beta_1_new = np.sum((Xtr_new-Xtr_new_mean)*(ytr_new-ytr_new_mean)) / np.sum((Xtr_new-Xtr_new_mean)**2)
beta_0_new = ytr_new_mean - beta_1_new*Xtr_new_mean
```

```python
Xts = np.arange(0,3.1,0.01)
print(Xts.shape)
```

```python
yhat = beta_1*Xts + beta_0 
ytrue = np.dot(transform(Xts), beta)
yhat_new = beta_1_new*Xts + beta_0_new
bias_2 = (yhat-ytrue)**2
```

```python
plt.figure(figsize=(10,5))
plt.plot(Xts, yhat, label="Estimated function train values in [0,1]")
plt.plot(Xts, yhat_new, label="Estimated function train values in [1,2]")
plt.plot(Xts, ytrue, label="True function")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.xlim([0,3])
plt.legend(loc="lower left")
plt.title("Estimated function vs True function", fontsize=12)
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.plot(Xts, bias_2, label="Bias^2")
plt.xlabel("x", fontsize=12)
plt.ylabel("Bias^2", fontsize=12)
plt.xlim([0,3])
plt.legend(loc="upper left")
plt.title("Bias^2 for training data in [0,1]", fontsize=12)
plt.grid()
```
